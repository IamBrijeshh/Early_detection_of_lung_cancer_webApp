# Data Augmentation

### Biomarker Column

```python
#import librery for gamma distribution
from scipy.stats import gamma

# Set shape and scale parameters
k = 2
theta = 3

# Create gamma distribution object
dist = gamma(k, scale=theta)

# Generate 10 random samples
samples = dist.rvs(size=309)

# Compute the cumulative distribution function
cdf = dist.cdf(samples)

# Transform the samples to be between 0 and 1
values = np.round(cdf,2)
print(values)

#convert values into 0 and 1 based on threshold value
#here threshold value = 0.5
biomarker_value = []
for value in values:
    if value > 0.5:
        value = 1
    else:
        value = 0
    print(value)
    biomarker_value.append(value)
```

## Model selection:

### Standardization of data

```python
#Separating the data and labels
X = df_new.values[:, 0:10]
Y = df_new.values[:,11] 
```

```python
X
```

### Split the Training and Testing data

```python
#Splitting data for training and testing
#train_test_split function will give 4 output X_train, X_test, y_train, y_test
#stratify = startifing based on y values

from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size = 0.3, stratify = Y, random_state = 10)
```

```python
print(X.shape,X_train.shape, X_test.shape)
```

### Train the model

```python
from sklearn import svm
svm_model = svm.SVC(kernel = 'linear')
```

```python
#training the SVM Classifier
svm_model.fit(X_train,y_train)
```

### Model Evaluation

#### Training Data

```python
#fitting training data to the model
X_train_prediction = svm_model.predict(X_train)

#accuracy score on training data
from sklearn.metrics import accuracy_score
train_data_accuracy = accuracy_score(X_train_prediction,y_train)
print("Accuracy Score on training data : ", train_data_accuracy , " %")
```

#### testing Data

```python
#fitting testing data to the model
X_test_prediction = svm_model.predict(X_test)

#accuracy score on testing data
test_data_accuracy = accuracy_score(X_test_prediction,y_test)
print("Accuracy Score on testing data : ", test_data_accuracy , " %")
```

```python
#Predicting result using testing data
y_pred= svm_model.predict(X_test)
y_pred
```

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
print(confusion_matrix(y_test,y_pred))
```

```python
#visualization
labels = [0,1]
cm = confusion_matrix(y_test, y_pred, labels=labels)

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
disp.plot();
```

## Predictive System

```python
input_data = (1,64,1,2,1,1,1,1,1,1)

# change the input data to a numpy array
input_data_as_numpy_array= np.asarray(input_data)

# reshape the numpy array as we are predicting for only on instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = svm_model.predict(input_data_reshaped)
print(prediction)

if (prediction[0]== 1):
  print('High Posibilites of lung cancer')
else:
  print('No risk of lung cancer')
```
